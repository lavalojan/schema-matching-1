```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from schemamatching.graph_generation import generate_graph, generate_pairwise_graph
```

```python
xml_file_path = 'data/countries_1.xml'
```

```python
with open(xml_file_path) as xml_file:
    xml_string = xml_file.read()
xml_string[:100]
```

```python
def display_graph(G):
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    nx.draw_networkx(G)
```

```python
G = generate_graph(xml_string)
display_graph(G)
```
